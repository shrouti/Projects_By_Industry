```python
#rename columns to easily know what they are
weather_df.rename(columns={'AWND': 'avg_wind_speed', 'PRCP': 'percipitation', 'SNWD': 'snow_depth','TSUN': 'total_sunshine','WT01': 'fog', 'WT02': 'heavy fog','WT03': 'thunder', 'WT04': 'sleet/small_hail','WT05': 'hail', 'WT06': 'glaze_or_rime','WT08': 'smoke_or_haze', 'WT09': 'heavy_snow'}, inplace=True)
```

```python
#changing name to airport code
abrv_name = []
for air in weather_df['NAME']:
    if air == 'LOS ANGELES INTERNATIONAL AIRPORT, CA US':
         abrv_name.append('LAX')
    elif air == 'JFK INTERNATIONAL AIRPORT, NY US':
         abrv_name.append('JFK')
    elif air == 'CHICAGO OHARE INTERNATIONAL AIRPORT, IL US':
         abrv_name.append('ORD')
    elif air == 'MIAMI INTERNATIONAL AIRPORT, FL US':
         abrv_name.append('MIA')
    else:
         abrv_name.append('DFW')
 
place = pd.Series((a for a in abrv_name))

weather_df['airport'] = place.values

weather_df = weather_df.drop(columns = "NAME")
```

```python
#seperating dates
weather_df['DATE'] = pd.to_datetime(weather_df['DATE'], format= "%Y/%m/%d" )

year = weather_df.DATE.map(lambda x: x.strftime('%-Y'))
month = weather_df.DATE.map(lambda x: x.strftime('%-m'))
day = weather_df.DATE.map(lambda x: x.strftime('%-d'))

weather_df['year'],weather_df['month'],weather_df['day'] = year.values.astype(int), month.values.astype(int), day.values.astype(int)

weather_df = weather_df.reindex(columns=['year','month','day','airport','TAVG', 'TMAX', 'TMIN', \
                           'percipitation', 'avg_wind_speed','SNOW','snow_depth','total_sunshine', \
                           'fog','heavy fog','thunder', 'sleet/small_hail','hail', 'glaze_or_rime','smoke_or_haze','heavy_snow','DATE'])
```
