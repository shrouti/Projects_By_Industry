```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
```

```python
flights = pd.read_csv('final_df.csv')
jfk_weather = pd.read_csv('JFK')
dfw_weather = pd.read_csv('DFW')
mia_weather = pd.read_csv('MIA')
ord_weather = pd.read_csv('ORD')
lax_weather = pd.read_csv('LAX')
```

```python
flights.head(3)
```

```python
jfk_weather['hours'] = pd.to_datetime(jfk_weather['datetime']).dt.hour
```

```python
dfw_weather['hours'] = pd.to_datetime(dfw_weather['datetime']).dt.hour
mia_weather['hours'] = pd.to_datetime(mia_weather['datetime']).dt.hour
ord_weather['hours'] = pd.to_datetime(ord_weather['datetime']).dt.hour
lax_weather['hours'] = pd.to_datetime(lax_weather['datetime']).dt.hour
```

```python
jfk_weather.head(3)
```

```python
jfk_new = jfk_weather[['datetime', 'New York_temp','New York_humidity', 'New York_wind', 'New York_pressure', 'New York','hours']]
```

```python
jfk_new.rename(columns = {'New York_temp': 'temp','New York_humidity': 'humidity', 'New York_wind': 'wind', 'New York_pressure': 'pressure', 'New York': 'description'}, inplace = True)
```

```python
ord_weather.rename(columns = {'Chicago_temp': 'temp','Chicago_humidity': 'humidity', 'Chicago_wind': 'wind', 'Chicago_pressure': 'pressure', 'Chicago': 'description'}, inplace = True)
```

```python
mia_weather.rename(columns = {'Miami_temp': 'temp','Miami_humidity': 'humidity', 'Miami_wind': 'wind', 'Miami_pressure': 'pressure', 'Miami': 'description'}, inplace = True)
dfw_weather.rename(columns = {'Dallas_temp': 'temp','Dallas_humidity': 'humidity', 'Dallas_wind': 'wind', 'Dallas_pressure': 'pressure', 'Dallas': 'description'}, inplace = True)
lax_weather.rename(columns = {'Los Angeles_temp': 'temp','Los Angeles_humidity': 'humidity', 'Los Angeles_wind': 'wind', 'Los Angeles_pressure': 'pressure', 'Los Angeles': 'description'}, inplace = True)
```

```python
mia_weather.drop(columns = 'Unnamed: 0', inplace = True)
dfw_weather.drop(columns = 'Unnamed: 0', inplace = True)
lax_weather.drop(columns = 'Unnamed: 0', inplace = True)
ord_weather.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
frame = [jfk_new, mia_weather, dfw_weather, ord_weather, lax_weather]
```

```python
jfk_new['airport'] = 'JFK'
```

```python
mia_weather['airport'] = 'MIA'
dfw_weather['airport'] = 'DFW'
lax_weather['airport'] = 'LAX'
ord_weather['airport'] = 'ORD'
```

```python
merged_weather = pd.concat(frame, axis = 0)
```

```python
merged_weather['Month'] = pd.to_datetime(merged_weather['datetime']).dt.month
merged_weather['DayofMonth'] = pd.to_datetime(merged_weather['datetime']).dt.day
```

```python
merged_weather.rename(columns = {'airport': 'Origin'}, inplace = True)
```

```python
merged_weather['hour'] = merged_weather['hours']
```

```python
flights.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
flights.head(3)
```

```python
flights['hour'] = flights['hour'].astype(int)
```

```python
flights.head(3)
```

```python
merged_flight_weather_df = pd.merge(flights, merged_weather, on = ['Month', 'DayofMonth', 'Origin', 'hour'])
```

```python
merged_flight_weather_df
```

```python
merged_flight_weather_df.to_csv('merged_flight_weather_df.csv')
```
